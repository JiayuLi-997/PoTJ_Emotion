### 3.1.2 & 3.1.3 bootstrap ratios

```python
def get_cohen(x1,x2):
    # calculate cohen's d
    x1_mean, x2_mean = x1.mean(), x2.mean()
    s1, s2 = x1.std(),x2.std()
    n1, n2 = len(x1), len(x2)
    s = np.sqrt( ( (n1-1)*s1**2 + (n2-1)*s2**2 ) / (n1+n2-2) )
    
    d = abs(x1_mean - x2_mean)/s
    se = np.sqrt( (n1+n2-1) / (n1+n2-3) * 4/(n1+n2)*(1+d**2/8) )
    CI = (d-1.96*se, d+1.96*se)
    
    return d, CI
```

```python
# valence

# Bootstrap for Fast 1000 times
Slow_all = singlev.loc[singlev.time=="Slow"]
Percentage = []
Slow_v = Slow_all.model_reweight_c.value_counts().sort_index()
for i in range(1000):
    Fast_sample = singlev.loc[singlev.time=="Fast"].sample(n=Slow_all.shape[0],random_state=i)
    Fast_v = Fast_sample.model_reweight_c.value_counts().sort_index()
    Percentage.append((Fast_v / Slow_v).to_numpy())
Percentage = np.array(Percentage)
mood_label = ["Negative","Neutral","Positive"]
Percentage_valence = Percentage.copy()

# test for bootstrapping results
mood_label=["negative","neutral","positive"]
p_values = np.zeros((len(mood_label),len(mood_label)))
t_sta = np.zeros((len(mood_label),len(mood_label)))
cohen_d = np.zeros((len(mood_label),len(mood_label))).astype(str)
for i in range(len(mood_label)):
    for j in range(i+1,len(mood_label)):
        p_values[i,j] = ttest_rel(Percentage_valence[:,i],Percentage_valence[:,j]).pvalue
        t_sta[i,j] = ttest_rel(Percentage_valence[:,i],Percentage_valence[:,j]).statistic
        d = get_cohen(Percentage_valence[:,i],Percentage_valence[:,j])
        cohen_d[i,j] = "%.2f (%.2f,%.2f)"%(d[0],d[1][0],d[1][1])

Percentage_valence_df = pd.DataFrame(Percentage_valence)
Percentage_valence_df.columns = mood_label
r = pd.DataFrame(p_values)*len(mood_label)*(len(mood_label)-1)/2
r.columns = mood_label
r.index=mood_label
print("t-test p value (corrected)")
display(r.round(4))

t_sta = pd.DataFrame(t_sta)
t_sta.columns, t_sta.index = mood_label,mood_label
t_sta[t_sta==0] = np.nan
print("t-test statistic")
display(t_sta.round(4))

print("cohen'd")
cohen = pd.DataFrame(cohen_d)
cohen.columns, cohen.index = mood_label,mood_label
cohen[cohen=='0.0'] = np.nan 
display(cohen)
```

```python
from scipy.interpolate import make_interp_spline
```

```python
# re-define the figure style
%load_ext autoreload
%autoreload 2
import matplotlib
matplotlib.rcParams['font.family'] = ['DejaVu Sans']
sns.set_style("white")
```

```python
def get_spline(x,y):
    m = make_interp_spline(x,y)
    xs = np.linspace(0,len(x)-1,100)
    ys = m(xs)
    return xs,ys

def get_between_fill(x,y,std):
    xs_e = np.linspace(0,len(x)-0.9,100)
    m_upper, m_lower = make_interp_spline(x,y+std), make_interp_spline(x,y-std)
    ys_upper, ys_lower = m_upper(xs_e), m_lower(xs_e)
    return xs_e, ys_lower, ys_upper
```

```python
# valence percentage changes
valence_day_percentage = singlev.groupby('date').model_reweight_c.value_counts(normalize=True).reset_index(name='percentage')
valence_day_percentage['month'] = valence_day_percentage.date.apply(lambda x: x[:7])
valence_month = valence_day_percentage.groupby(['month','model_reweight_c']).percentage.agg(['mean','std']
                                                                                           ).reset_index()
positive_avg = valence_month.loc[valence_month.model_reweight_c==1].sort_values(by='month')
negative_avg = valence_month.loc[valence_month.model_reweight_c==-1].sort_values(by='month')

plt.figure(figsize=(10,5))
ax = plt.subplot()
x = np.arange(len(positive_avg))
y1 = positive_avg['mean'].to_numpy()
y2 = negative_avg['mean'].to_numpy()
xs, ys1 = get_spline(x,y1)
xs, ys2 = get_spline(x,y2)

plt1, = ax.plot(xs,ys1,color='tab:red')
ax.plot(x,y1,marker='s',linestyle='',color='tab:red')
plt2, = ax.plot(xs,ys2,color='tab:blue')
ax.plot(x,y2,marker='o',linestyle='',color='tab:blue')
ax.legend([plt1,plt2],['positive','negative'],fontsize=20,loc=0)

# fill standard divation area
xs_e,ys1_lower,ys1_upper = get_between_fill(x,y1,positive_avg['std'].to_numpy())
ax.fill_between(xs_e,ys1_lower,ys1_upper,facecolor='tab:red',alpha=0.3)
xs_e,ys2_lower,ys2_upper = get_between_fill(x,y2,negative_avg['std'].to_numpy())
ax.fill_between(xs_e,ys2_lower,ys2_upper,facecolor='tab:blue',alpha=0.3)

positive_range = ys1_upper.max()-ys1_lower.min()
negative_range = ys2_upper.max()-ys2_lower.min()
positive_mean, negative_mean = positive_avg['mean'].mean(),negative_avg['mean'].mean()
ax.plot([0,len(positive_avg)-1],[positive_mean,positive_mean],linestyle='--',color='tab:red')
ax.plot([0,len(positive_avg)-1],[negative_mean,negative_mean],linestyle=':',color='tab:blue')
ax.set_ylim(positive_mean-positive_range*0.8,positive_mean+positive_range*0.8)

ax.set_xticks(np.arange(len(positive_avg)))
ax.set_xticklabels([str(x)[5:] for x in positive_avg.month],fontsize=19)
ax.set_ylabel("Avg. Percentage\nof Daily Tweets",fontsize=22)
ax.set_xlabel("Month",fontsize=22)
plt.savefig(os.path.join(figdir,"avgday_distribution_valence_per.pdf"),bbox_inches='tight')
```
